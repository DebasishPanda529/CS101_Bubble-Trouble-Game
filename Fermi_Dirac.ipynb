```python

import numpy as np
import matplotlib.pyplot as plt 

#Physical Constants
kB_J=1.3806503e-23 #Boltzmnann constant in J.K^-1
q=1.6e-19 #Electron charge in Coulomb 

#Boltzmann constant in eV
kB=kB_J/q #Boltzmnann constant in eV.K^-1

N=100
Ef=0 #Assuming Fermi energy to be zero, chossing as a reference
T1=50
T2=300 #Tempereature in K
T3=600

E,dE=np.linspace(-0.4,0.4,num=N,retstep=True,endpoint=True)

#Computation of Fermi-Dirac Statistics
f1=1/(1+np.exp((E-Ef)/(kB*T1)))
f2=1/(1+np.exp((E-Ef)/(kB*T2)))
f3=1/(1+np.exp((E-Ef)/(kB*T3)))

#Plot the graph
plt.plot(E,f1,'g-',label="T1=50K")
plt.plot(E,f2,'r-',label="T2=300K")
plt.plot(E,f3,'b-',label="T3=600K")
plt.title("Fermi-Dirac Statistics")
plt.xlabel("Energy (eV)")
plt.ylabel("f(E)")
plt.legend(loc="center left")
plt.show()
```
